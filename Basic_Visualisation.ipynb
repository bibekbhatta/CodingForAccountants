# This tutorial is to produce simple graphs using matplotlib library in python. Matplotlib is a polular visualization tool.

For full tutorial on matplotlib, go to https://matplotlib.org/stable/index.html

```python
import matplotlib.pyplot as plt # import the necessary package and call it plt
```

If you want to change the style of the graphs (e.g. background colours), uncomment the below two codes and choose the style you prefer. Otherwise, proceed to the next cell. 

```python
# print(plt.style.available) # displays available styles
# plt.style.use('classic') # to choose style, select from available styles 
```

```python
X = [1,2,3,4,5] # setting the variables
Y = [1,2,3,4,10]
Z = [2,3,4,5,11]
```

# plot Y and Z against X in one figure

```python
plt.figure(figsize=(10,7)) # width =10; height =7
plt.plot(X, Y, 'bo-', label='Blue with dots and straight line')  #b=blue ; o=dots; - = straight line
plt.plot(X, Z, 'g*', label='Green with stars')  # g=green; *=stars
plt.title('Two Scatterplots in one figure')  
plt.xlabel('X variables')
plt.ylabel('Y variables')
plt.xlim(0, 6) # x axis from 0 to 6
plt.ylim(0, 12) # y axis from 0 to 12
plt.legend(loc='best')# chooses the best location for legend
plt.show()
```

# Plot the two graphs in separately

```python
### Create 1 x 2 grid; with y axis shared between the two graphs
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,4), sharey=True)
# Plot
ax1.plot(X, Y, 'b.-')  # blue dots and a line
ax2.plot(X, Z, 'g*')  # green stars; no line
ax1.set_title('Scatterplot (blue small dots + line)'); ax2.set_title('Scatterplot (green stars)')
ax1.set_xlabel('X var');  ax2.set_xlabel('X green')  # x label
ax1.set_ylabel('Y var');  ax2.set_ylabel('Y var')  # y label
ax1.set_xlim(0, 6) ;  ax2.set_xlim(0, 6)   # both x axis set from 0 to 6
ax1.set_ylim(0, 12);  ax2.set_ylim(0, 12)  # y axis from 0 to 12
# ax2.yaxis.set_ticks_position('none') 
plt.tight_layout()
plt.show()
```
