# Artificial sales data from three cities

```python
import numpy as np # importing the number-crunching package called numpy and giving it a name 'np'

salesA = np.random.randint(5,150, size=(5000000,1)) #create random sales data for City A
salesB = np.random.randint(45,175, size=(5000000,1)) #create random sales data for City B
salesC = np.random.randint(140,450, size=(5000000,1)) #create random sales data for City C

salesAll = np.concatenate((salesA,salesB,salesC), axis=1) # sales of three cities brought together side by side (Axis=1)


```

```python
import pandas as pd #importing pandas package for data manippulation and giving it a name 'pd'
data = pd.DataFrame(salesAll, columns=['CityA','CityB','CityC']) # bringing the sales data into pandas table
```

```python
data.plot.box() # visual/box plot for sales data
```

```python
data.describe() # numerical summary of the sales data
```

```python
data.CityC.value_counts() # frequency table for sales amount for CityC
```
